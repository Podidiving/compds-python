```python
class MyPrettyInt:
    def __init__(self, val: int):
        self.val = str(val)

    def __repr__(self):
        return f"MyPrettyInt(val={self.val})"

    def __str__(self):
        return  "__str__ " + self.val
```

```python
obj = MyPrettyInt(123)

print(obj)
```

```python
str(obj)
```

```python
obj.__repr__()
```

Напишите функцию, которая:

- получает на вход строку (возможно, небольшой текст) и стоп-слово
- находит все индексы стоп-слова в строке
- в результате своей работы собирает новую строку, в которой нет стоп-слов

```python
def func(inp: str, stop: str) -> str:
    return inp.replace(stop, "")

input_str = "Hello stapworld stap!"
stop = "stap"
output_str = "Hello world !"


assert func(input_str, stop) == output_str
```

```python
def func2(inp: str, stop: str) -> str:
    while stop in inp:
        index = inp.find(stop)
        inp = inp[:index] + inp[index + len(stop):]
    return inp

    # inp.find ....
    # for / while
    # slice


input_str = "Hello stapworld stap!"
stop = "stap"
output_str = "Hello world !"


assert func2(input_str, stop) == output_str
```

```python
# walrus operator
def func3(inp: str, stop: str) -> str:
    # python3.9 (3.8?)
    while (index := inp.find(stop)) != -1:
        inp = inp[:index] + inp[index + len(stop):]
    return inp

input_str = "Hello stapworld stap!"
stop = "stap"
output_str = "Hello world !"


assert func2(input_str, stop) == output_str
```
